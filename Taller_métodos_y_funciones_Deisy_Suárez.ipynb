# FUNCIONES Y METODOS EN LISTAS

```python
#Declaramos la lista Medicamentos

Medicamentos=["Antibióticos", "Antipiréticos", "Antialérgicos", "Antimicóticos", "Corticoides"]
print("Estos son los tipos de medicamentos que tenemos en inventario",  Medicamentos)
```

```python
#len(lista): Devuelve el número de elementos en la lista.

print("Cantidad de Tipos de Medicamentos",len(Medicamentos))
```

```python
#sum(lista): Retorna la suma de todos los elementos numéricos de la lista.

cantidad_por_medicamentos=[10 ,15, 20, 25,30]
print("Cantidad de Medicamentos",sum(cantidad_por_medicamentos))
```

```python
#max(lista): Devuelve el elemento más grande de la lista.

print("Cantidad máxima de Medicamentos",max(cantidad_por_medicamentos))
```

```python
#min(lista): Devuelve el elemento más pequeño de la lista.

print("Cantidad mínima de Medicamentos",min(cantidad_por_medicamentos))
```

```python
#sorted(lista): Retorna una nueva lista ordenada sin modificar la original.

print("Medicamentos Ordenados",sorted(Medicamentos))
```

```python
#enumerate()– Permite recorrer la lista obteniendo índices y valores

for indice,valor in enumerate(Medicamentos):
    print(indice,valor)
```

```python
#Función Zip combina dos listas Ej: Medicamentos y cantidad por medicamentos

cantidad_tipo_medicamentos=list(zip(Medicamentos,cantidad_por_medicamentos))
print("Cantidad por tipo de Medicamentos",cantidad_tipo_medicamentos)
```

```python
#lista.append(elemento): Agrega un elemento al final de la lista.

Medicamentos.append("Antidepresivos")
print("Medicamentos con Antidepresivos",Medicamentos)
```

```python
# lista.extend: Agrega múltiples elementos al final.

Medicamentos.extend(["Analgésicos","Antihistamínicos"])
print("Medicamentos con Analgésicos y Antihistamínicos",Medicamentos)
```

```python
#lista.insert(posición, elemento): Inserta un elemento en una posición específica.

Medicamentos.insert(2,"Antidepresivos")
print("Medicamentos con Antidepresivos",Medicamentos)
```

```python
#lista.remove(elemento): Elimina la primera aparición del elemento en la lista.

Medicamentos.remove("Antidepresivos")
print("Medicamentos sin Antidepresivos",Medicamentos)
```

```python
#lista.sort: Ordena la lista en su lugar.

Medicamentos.sort()
print("Medicamentos Ordenados",Medicamentos)
```

#FUNCIONES Y METODOS EN CONJUNTOS (SET)

```python
#SET

Presentación={"cápsulas", "píldoras", "tabletas", "soluciones", "polvos", "granulados", "cremas"}
print("Presentación de Medicamentos",Presentación)
```

```python
#conjunto.add(elemento): Agrega un elemento al conjunto.

Presentación.add("suspensión")
print("Presentación de Medicamentos",Presentación)
```

```python
#conjunto.remove(elemento): Elimina un elemento, pero genera error si no existe

#conjunto.remove(elemento): Elimina un elemento, pero genera error si no existe

if "gotas" in Presentación:
  Presentación.remove("gotas")
else:
  print("No existe la presentacion de gotas")
print("Presentación de Medicamentos",Presentación)
```

```python
#conjunto.discard(elemento): Elimina un elemento sin generar error si no existe.

Presentación.discard("gotas")
print("Presentación de Medicamentos",Presentación)
```

```python
#set.union(otro_set): Retorna un nuevo conjunto con la unión de ambos.

Otras_Presentaciones={"geles", "aerosoles","tabletas", "nebulizadores","cremas", "jarabes", "gotas"}
print("Presentación de Medicamentos",Presentación.union(Otras_Presentaciones))
```

```python
#set.intersection(otro_set): Retorna la intersección de ambos conjuntos.

print("Presentación en ambos conjuntos",Presentación.intersection(Otras_Presentaciones))
```

```python
#set.difference(otro_set): Retorna la diferencia (elementos presentes en el primero pero no en el segundo).

print("Elementos incluidos solo en un conjunto",Presentación.difference(Otras_Presentaciones))
```

```python
#set.symmetric_difference(otro_set): Retorna los elementos que están en uno u otro conjunto, pero no en ambos.

print("Elementos que no están en ambos conjuntos",Presentación.symmetric_difference(Otras_Presentaciones))
```

#FUNCIONES Y METODOS CON TUPLAS

```python
#Podemos acceder a los elementos de una tupla mediante índices

Códigos=(1001,1002,1003,1004,1005,1006,1007)
print("Devuelve el primer elemento",Códigos[0])
print("Devuelve el segundo elemento",Códigos[1])
print("Devuelve el último elemento", Códigos[-1])
```

```python
#count(elemento): Cuenta cuantas veces aparece un elemento.

print("Cantidad de veces que aparece el código 1001=",Códigos.count(1001))
```

```python
#index(elemento): Devuelve el índice de la primera aparición del elemento

print("Indice del código 1002=",Códigos.index(1002))
```

```python
#Convertir Tupla a Lista, se usa para poder modificar los datos

Lista_de_Códigos=list(Códigos)
print("Lista de Códigos",Lista_de_Códigos)

#Ahora podemos agregar datos a la nueva lista

Lista_de_Códigos.append(1008)
print("Lista de Códigos actualizada",Lista_de_Códigos)
```

#FUNCIONES Y METODOS PARA TRABAJAR CON DICCIONARIOS


```python
#Diccionarios que se utilizarán en este ejercicio

Medicamento_y_Presentación={"Antibióticos":"Tabletas", "Antipiréticos":"Píldoras", "Antialérgicos":"Cápsulas", "Antimicóticos":"Cremas", "Corticoides":"Suspensión"}
Otros_Medicamentos={"Analgésicos":"Tabletas", "Antihistamínicos":"Cápsulas"}
Código_Medicamento={1001:"Antibióticos", 1002: "Antipiréticos", 1003: "Antialérgicos", 1004: "Antimicóticos", 1005:"Corticoides", 1006:"Analgésicos", 1007:"Antihistamínicos"}
print("Medicamento y presentación",Medicamento_y_Presentación,Otros_Medicamentos)
print("Código de Medicamento",Código_Medicamento)
```

```python
#len(diccionario) Devuelve el número de elementos en el diccionario.

print("Cantidad de Medicamentos",len(Medicamento_y_Presentación))
```

```python
#max(diccionario) y min(diccionario) Devuelven la clave máxima y mínima del diccionario (según orden alfabético o numérico de las claves).

print(max(Código_Medicamento))
print(min(Código_Medicamento))
```

```python
#keys() Devuelve una vista de todas las claves.

print("Códigos",Código_Medicamento.keys())
```

```python
#values() Devuelve una vista de todos los valores.

print("Medicamentos",Código_Medicamento.values())
```

```python
#items()  Devuelve una vista de los pares clave-valor en tuplas.

print("Código - Medicamento",Código_Medicamento.items())
```

```python
#get(clave, valor_por_defecto) Obtiene un valor dado su clave. Si no existe, devuelve el valor por defecto

print("Presentación del Medicamento Antibióticos=",Medicamento_y_Presentación.get("Antibióticos","No existe en este directorio"))
print("Presentación del Medicamento Analgésicos=",Medicamento_y_Presentación.get("Analgésicos","No existe en este directorio"))
```

```python
#update(otro_diccionario) Fusiona otro diccionario dentro del actual, sobrescribiendo claves repetidas.

Medicamento_y_Presentación.update(Otros_Medicamentos)
print("Medicamento y presentación",Medicamento_y_Presentación)
```

```python
#setdefault(clave, valor_por_defecto)  Si la clave existe, devuelve su valor. Si no, la agrega con un valor por defecto.

print("Presentación del Medicamento Analgésicos=",Medicamento_y_Presentación.setdefault("Analgésicos","No existe"))
print("Presentación del Medicamento Aspirinas=",Medicamento_y_Presentación.setdefault("Aspirinas","No existe"))
print("Medicamento y presentación",Medicamento_y_Presentación)
```
